```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import tclab

print("Version = ", tclab.__version__)
tclab.diagnose()
```

```python
# Exercise 1

# Define functions for u1(t) and u2(t) (using interpolation) that create the input signals

t1_expt = [20,20,1020,1020]

u1_expt = [0,60,60,0]

t2_expt = [620,620,1420,1420]

u2_expt = [0,60,60,0]

def u1(t):
    return np.interp(t, t1_expt, u1_expt)

def u2(t):
    return np.interp(t, t2_expt, u2_expt)

time=np.linspace(-100,1800,1901)

plt.plot(time,u1(time))
plt.plot(time,u2(time))
plt.title('Input Signals')
plt.xlabel('Time (seconds)')
plt.ylabel('% Max Power')
```

```python
# Exercise 2

# Collect the experimental data and plot the results

# Note: I added 100 extra seconds at the beginning of the experiment (so the time range is from 0 to 1900 seconds total)
# to account for the initial "-100" seconds that the device is at rest with the heaters turned off, before the experiment begins 
# at time 0, or in this case, at t=100.
 
%matplotlib inline

from tclab import TCLab, clock, Historian, Plotter

with TCLab() as lab:
    lab.P1=200
    lab.P2=100
    h = Historian(lab.sources)
    p = Plotter(h, 1900)
    for t in clock(1900):
        x=t-100 
        lab.Q1(0 if t<=100 else u1(x))
        lab.Q2(0 if t<=100 else u2(x))
        p.update(t)
   
        
# save the data
h.to_csv('data_lab2_1.csv')
```
